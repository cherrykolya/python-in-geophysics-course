```python
# Код для загрузки данных


import httpx

# URL с CSV-данными Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Имя файла для сохранения
file_name = "titanic.csv"

# Скачивание данных
response = httpx.get(url)

if response.status_code == 200:  # Проверяем успешность запроса
    with open(file_name, "wb") as file:
        file.write(response.content)
    print(f"Датасет успешно скачан и сохранен в '{file_name}'.")
else:
    print(f"Ошибка при скачивании: {response.status_code}")
```
